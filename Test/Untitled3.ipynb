```r
for(i in 1:nrow(df_screen)){
    temp_actors <- unlist(strsplit(df_screen$배우[i],","))
    for(j in 1:nrow(ju_df)){
        if(ju_df$주연[j] %in% temp_actors){
            ju_df$횟수[j] = ju_df$횟수[j] + 1
            ju_df$누적관객수__개봉일스크린수[j] = ju_df$누적관객수__개봉일스크린수[j] +df_screen$누적관객수__개봉일스크린수[i]
            ju_df$누적관객수__스크린수[j] = ju_df$누적관객수__스크린수[j] + df_screen$누적관객수__스크린수[i]
            ju_df$누적관객수__상영횟수[j] = ju_df$누적관객수__상영횟수[j] + df_screen$누적관객수__상영횟수[i]
        }
    }
}
```

```r
ju_df$평균_관람객_개스크 <- ju_df$누적관객수__개봉일스크린수/ju_df$횟수
ju_df$평균_관람객_스크 <-ju_df$누적관객수__스크린수/ju_df$횟수
ju_df$평균_관람객_상횟  <-ju_df$누적관객수__상영횟수/ju_df$횟수
```
