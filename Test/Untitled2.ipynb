```r
df<- read.csv("../movies/temp_mining_ver1.csv", stringsAsFactors = F)
actor_score_df <- read.csv("../movies/actor_score_ver2.csv", stringsAsFactors = F)
```

```r
library(dplyr)
```

```r
df$배우누적점수 <-0
df$배우점수 <- 0
df$배우수 <- 0
```

```r

for(i in 1:nrow(df)){
    score<-c()
    num <- 0
    for(j in 1:nrow(actor_score_df)){
        if( actor_score_df$주연[j] %in% unlist(strsplit(df$주연[i],",")) ){
            num = num + 1
            score = c(score,actor_score_df$평균점수[j])
        }
    }
        if(length(score) != 0){
        df$배우점수[i] = max(score)
        df$배우누적점수[i] = sum(score)
        df$배우수[i] = num        
    }

    

}
```

```r
cor(df$누적관객수, df$배우점수)
```
