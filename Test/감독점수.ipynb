```r
df <- read.csv("../movies/movie109_ver1.csv",fileEncoding = "utf-8")

library(dplyr)

df <- arrange(df , desc(누적관객수))

head(df)



str(df)

df$감독 <- as.character(df$감독)

ju <- c()

for (i in 1: nrow(df)){
    temp_ju <- unlist(strsplit(df$감독[i],","))
    ju <- c(ju,temp_ju)
}

ju_df <- data.frame(table(ju))
colnames(ju_df) <- c('감독',"횟수")

head(ju_df)

ju_df$점수 = 0
head(ju_df)

df$감독 <- as.character(df$감독)

for(i in 1:nrow(df)){
    for( j in 1:nrow(ju_df)){
        if(ju_df[j,1] %in% unlist(strsplit(df$감독[i],","))){
            if(df$cluster[i] == 1){
                ju_df$점수[j] = ju_df$점수[j] + 1
            }else if(df$cluster[i] == 2){
                ju_df$점수[j] = ju_df$점수[j] + 2
            }else{
                ju_df$점수[j] = ju_df$점수[j] + 4
            }
        }
    }
}

ju_df<- arrange(ju_df,desc(점수))

head(ju_df,10)
```

```r
ju_df$평균점수 <- ju_df$점수 / ju_df$횟수
```

```r

temp_df <- subset(arrange(ju_df,desc(평균점수)),횟수 >=2)
```

```r
write.csv(temp_df, "../movies/director_score_109.csv",row.names = F)
```
