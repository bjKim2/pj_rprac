```r
dir_actor_df <- read.csv("../movies/감독_배우_점수.csv",fileEncoding = 'utf-8' , stringsAsFactors = F)
df <- read.csv("../movies/over200mil_change_actor.csv",fileEncoding = 'utf-8', stringsAsFactors =  F )
```

```r
library(dplyr)
```

```r
for(i in 1:nrow(df)) 
df$조연[i]<- substr(df$조연[i],2,nchar(df$조연[i]))
```

```r
df$감독_주연시너지수 = 0
df$감독_주연악연수 = 0
```

```r
head(df)
```

```r
df$감독_주연시너지수 = 0
df$감독_주연악연수 = 0
for( i in 1:nrow(df)){
    df_dir = unlist(strsplit(df$감독[i], ","))
    df_actor = unlist(strsplit(df$주연[i], ","))
    for( j in 1:nrow(dir_actor_df)){
        temp_da = unlist(strsplit(dir_actor_df$감독_배우[j], "_"))
        if(temp_da[1] %in% df_dir & temp_da[2] %in% df_actor){
            if(dir_actor_df[j,2] > 0){
                df$감독_주연시너지수[i] = df$감독_주연시너지수[i] + 1
            }else if(dir_actor_df[j,2] < 0){
                df$감독_주연악연수[i] = df$감독_주연악연수[i] + 1
            }
        }
    }
}
```

```r
tail(df)
```

```r
df$감독_조연시너지수 = 0
df$감독_조연악연수 = 0
for( i in 1:nrow(df)){
    df_dir = unlist(strsplit(df$감독[i], ","))
    df_actor = unlist(strsplit(df$조연[i], ","))
    for( j in 1:nrow(dir_actor_df)){
        temp_da = unlist(strsplit(dir_actor_df$감독_배우[j], "_"))
        if(temp_da[1] %in% df_dir & temp_da[2] %in% df_actor){
            if(dir_actor_df[j,2] > 0){
                df$감독_조연시너지수[i] = df$감독_조연시너지수[i] + 1
            }else if(dir_actor_df[j,2] < 0){
                df$감독_조연악연수[i] = df$감독_조연악연수[i] + 1
            }
        }
    }
}
```

```r
cor(df$누적관객수, df$감독_주연시너지수)
cor(df$누적관객수, df$감독_주연악연수)
cor(df$누적관객수, df$감독_조연시너지수)
cor(df$누적관객수, df$감독_조연악연수)

cor(df$누적관객수, df$감독_주연시너지수-df$감독_주연악연수)
cor(df$누적관객수, df$감독_조연시너지수-df$감독_조연악연수)

cor(df$누적관객수, df$감독_주연시너지수+df$감독_조연시너지수)
cor(df$누적관객수, df$감독_주연시너지수-df$감독_주연악연수+df$감독_조연시너지수-df$감독_조연악연수)
```

```r
par(mfrow = c(2,2))
plot(df$누적관객수 ~ df$감독_주연시너지수)
plot(df$누적관객수 ~ df$감독_주연악연수)
plot(df$누적관객수 ~ df$감독_조연시너지수)
plot(df$누적관객수 ~ df$감독_조연악연수)

plot( (df$감독_주연시너지수+df$감독_조연시너지수),df$누적관객수 )
```
