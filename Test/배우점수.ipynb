```r
ju_df$점수 <- 0
for(i in 1:nrow(df)){
    for( j in 1:nrow(ju_df)){
        if(ju_df[j,1] %in% unlist(strsplit(df$주연[i],","))){
            if(df$cluster[i] == 1){
                ju_df$점수[j] = ju_df$점수[j] + 1
            }else if(df$cluster[i] == 2){
                ju_df$점수[j] = ju_df$점수[j] + 2
            }else if(df$cluster[i] == 3){
                ju_df$점수[j] = ju_df$점수[j] + 3
            }else if(df$cluster[i] == 4){
                ju_df$점수[j] = ju_df$점수[j] + 4
            }else{
                ju_df$점수[j] = ju_df$점수[j] + 5
            }
        }
    }
}
```
