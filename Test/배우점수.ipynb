```r
df$주연 <- as.character(df$주연)
df$조연 <- as.character(df$조연)
```

```r
ju <- c()

for (i in 1: nrow(df)){
    temp_ju <- unlist(strsplit(df$주연[i],","))
    ju <- c(ju,temp_ju)
}
ju_df <- data.frame(table(ju))
colnames(ju_df) <- c('주연',"횟수")

head(ju_df)
```

```r
ju_df$점수 = 0
head(ju_df)
```

```r
for(i in 1:nrow(df)){
    for( j in 1:nrow(ju_df)){
        if(ju_df[j,1] %in% unlist(strsplit(df$주연[i],","))){
            if(df$cluster[i] == 1){
                ju_df$점수[j] = ju_df$점수[j] + 1
            }else if(df$cluster[i] == 2){
                ju_df$점수[j] = ju_df$점수[j] + 4
            }else if(df$cluster[i] == 3){
                ju_df$점수[j] = ju_df$점수[j] + 9
            }else if(df$cluster[i] == 4){
                ju_df$점수[j] = ju_df$점수[j] + 16
            }else{
                ju_df$점수[j] = ju_df$점수[j] + 25
            }
        }
    }
}
```

```r
ju_df<- arrange(ju_df,desc(점수))
```

```r
head(ju_df,10)
```

```r
ju_df$평균점수 = 0
```

```r
for( i in 1: nrow(ju_df)){
    ju_df$평균점수[i] <- ju_df$점수[i] / ju_df$횟수[i] 
}
```
