```r
for(i in 1:nrow(dir_df)){
    if(dir_df$누적관객수[i] < 2895508){
        dir_df$cluster[i] = 1 
    }else if(dir_df$누적관객수[i] < 4382976){
        dir_df$cluster[i] = 2
    }else if(dir_df$누적관객수[i] < 6514330){
        dir_df$cluster[i] = 3
    }else if(dir_df$누적관객수[i] < 10085275){
        dir_df$cluster[i] = 4
    }else{
        dir_df$cluster[i] = 5
    }
}
```
