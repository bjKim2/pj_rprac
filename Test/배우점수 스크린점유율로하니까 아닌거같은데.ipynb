```r
df<-read.csv("../movies/ju_score.csv", fileEncoding = 'utf-8')
```

```r
head(df)
```

```r
df$최종점수 <- df$평균_관람객_개스크 * (df$횟수)^(1/2)  * df$cluster^2
```

```r
hist(subset(df, df$cluster == 4)$최종점수)
```

```r
options(scipen = 100)
par(mfrow = c(3,3))
plot(df$cluster,df$최종점수)
plot(df$cluster,df$누적관객수__개봉일스크린수)       
plot(df$cluster,df$누적관객수__스크린수)  
plot(df$cluster,df$누적관객수__상영횟수)  
plot(df$cluster,df$평균_관람객_개스크)  
plot(df$cluster,df$평균_관람객_스크)  
plot(df$cluster,df$평균_관람객_상횟)
plot(df$cluster,df$관람객개스크점수)
```

```r
cor(df$평균_관람객_개스크,df$횟수)
plot(b,a)
```
