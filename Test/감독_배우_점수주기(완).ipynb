```r
dir_actor_df <- data.frame(감독_배우 = unique( dir_df$감독_배우) , 점수 = 0)
```

```r
head(dir_actor_df)
```

```r
for(i in 1:nrow(dir_actor_df)){
    temp_df<- dir_df[dir_actor_df$감독_배우[i] == dir_df$감독_배우, ]
    for(j in 1:nrow(temp_df)){
        if (temp_df$등급[j] == '상'){
            dir_actor_df$점수[i] = dir_actor_df$점수[i] +1
        }else if(temp_df$등급[j] == '하'){
            dir_actor_df$점수[i] = dir_actor_df$점수[i] -1
        }
    }
}
```
