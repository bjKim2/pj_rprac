```r
df<-read.csv("./movies/new_movie_ver3.csv")
```

```r
df$감독 <- as.character(df$감독)
```

```r
x<-c()
for(i in 1:nrow(df)){
    temp <-unlist(strsplit(df$감독[i],","))
    if(length(temp)>=2){
        for(i in 1:length(temp)){
            x<- c(x,temp[i])
        }
    }else{
        x<- c(x,temp)
    }
}
```

```r
library(dplyr)
```

```r
temp_df <- as.data.frame(table(x))
temp_df<- arrange(temp_df,desc(Freq))
```

```r
colnames(temp_df)<-c("감독","횟수")
```

```r
write.csv(temp_df,"./movies/directors.csv",row.names=F)
```

```r
temp_df$점수 <- 0

for (j in 1:nrow(temp_df)) {
    temp <- c()
    for (i in 1:nrow(df)) {
        if(temp_df[j,1] %in% unlist(strsplit(df$감독[i],","))) {
            if(df$cluster[i] == 3){
                temp_df[j,3] = temp_df[j,3] + 3
            }else if(df$cluster[i] == 2){
                temp_df[j,3] = temp_df[j,3] + 1
            }else{
                temp_df[j,3] = temp_df[j,3] + 2
            }
            
        }
    }
}
```

```r
director_rank <- arrange(temp_df,desc(점수))
```
