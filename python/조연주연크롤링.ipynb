```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
data = pd.read_csv("../movies/over200mil.csv")


data['배우']  = ''
data['조연']= ''

for i in range(0,len(data)):
    
    check = False
    index = 100
    actors = []
    print("회차 : "+str(i))
    
    for o in range(1,6):
        url = 'https://movie.naver.com/movie/search/result.naver?section=movie&query=' + data['영화명'][i] + '&page=' + str(o)
        response = requests.get(url)
    
        if response.status_code == 200:
            html = response.text
            soup = BeautifulSoup(html, 'html.parser')

            title = soup.select('#old_content > ul:nth-child(4) > li ')

            sub_title = soup.select("#old_content > ul:nth-child(4) > li > dl > dt > a ")

            print(str(len(title)) + " 타이틀 수")
            for k in range(0,len(title)):
                director = soup.select('#old_content > ul.search_list_1 > li:nth-child('+str(k+1)+') > dl > dd:nth-child(4) > a:nth-child(1)')

                if(len(director) > 0):
                    if(director[0].text in data['감독'][i].split(",") ):
                        index = k
                        print(director[0].text + "감독체크")
                        print("index : " + str(index))

                        break
            if(index == 100):
                data['배우'][i] = ""
                continue

            bc = sub_title[index].attrs

            new_url = "https://movie.naver.com" + bc["href"].split(".")[0][0:13] + "detail." + bc["href"].split(".")[1]

            print(new_url)

            response = requests.get(new_url)

            if response.status_code == 200:
                html = response.text
                soup = BeautifulSoup(html, 'html.parser')

                actor = soup.select("#content > div.article > div.section_group.section_group_frst > div.obj_section.noline > div > div.lst_people_area.height100 > ul > li > div > a")
                grade = soup.select("#content > div.article > div.section_group.section_group_frst > div.obj_section.noline > div > div.lst_people_area.height100 > ul > li > div > div > p.in_prt > em")


                data['배우'][i] = data['배우'][i] + actor[0].text
                for j in range(1,len(actor)):
                    if(grade[j].text == "주연"):
                        data['배우'][i] = data['배우'][i] + ","+ actor[j].text
                    else:
                        data['조연'][i] = data['조연'][i] + ","+ actor[j].text
            break

    
    else : 
        print(response.status_code)
```
