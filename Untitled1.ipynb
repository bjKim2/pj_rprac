```r
library(dplyr)
```

```r
fun_actor <- function(x){
    options(scipen = 100)
    options(warn = -1)
    par(family = "AppleGothic")
    par(mfrow = c(2,2))

    temp_df <-   subset(x = df,배우 == x,select = c(배우,개봉일,누적관객수))
    temp_df$배우 <- as.character(temp_df$배우)

    temp_df$배우 <- as.factor(temp_df$배우)

    rownames(temp_df) <- c()
    temp_df <- arrange(temp_df,누적관객수)
    plot(
        rownames(temp_df),
        temp_df[,"누적관객수"],    
        type= "o",
    )
    plot(
        temp_df[,2],
        temp_df[,"누적관객수"],    
        type= "o",
    )
    hist(
        temp_df[,"누적관객수"],    
    )
    
}
fun_actor('송강호')
```

```r
fun_director <- function(x){
    options(scipen = 100)
    options(warn = -1)
    par(family = "AppleGothic")
    par(mfrow = c(2,2))

    temp_df <-   subset(x = df,감독 == x,select = c(감독,배우,개봉일,누적관객수))
    temp_df$감독 <- as.character(temp_df$감독)

    temp_df$감독 <- as.factor(temp_df$감독)

    rownames(temp_df) <- c()
    temp_df <- arrange(temp_df,누적관객수)
    plot(
        rownames(temp_df),
        temp_df[,"누적관객수"],    
        type= "o",
    )
    plot(
        temp_df[,3],
        temp_df[,"누적관객수"],    
        type= "o",
    )
    hist(
        temp_df[,"누적관객수"],    
    )

}
fun_director('엠씨짱')
```
