```r
df2$주연 <- as.character(df2$주연)
temp_df <- c()

for(i in 1:nrow(df2)){
    temp <- c()
    x <- unlist(strsplit(df2$주연[i],","))
    for(j in 1:length(x)){
        temp = rbind(temp,data.frame(영화명 = df2$영화명[i] , 주연 = x[j] ))
    }
    temp_df <- rbind(temp_df,temp)
}
    
```
