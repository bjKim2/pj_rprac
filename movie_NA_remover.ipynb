```r
df[df$영화명== '빛나는 거짓',"등급"] <- '15세이상관람가'
#등급칸만 NA 인 영화에 검색해서 데이터 삽입

rownames(df) <-c()

x <- c()
for (i in 1: nrow(df)){
    if( df$등급[i]== ""){
       x <- c(x,i) 
    }
}
# 등급이 NA 인 샘플 제거
x


df <- df[-x,]
#제거완료

head(df)
nrow(df)
rownames(df)<-c()

for(i in 1:ncol(df))
    print(sum(df[,i] == ""))
```
