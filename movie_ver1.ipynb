```r
#배우를 제외한 데이터 NA 처리 완료,(배우쪽은 애니메이션 다큐멘터리의 경우 배우가 없는 경우가 존재하기 때문에 아직 처리하지 않음),
# 배우,감독,배급사 등 컬럼값 중 다수 데이터 존재시 맨 앞 값으로 수정
#NA 처리 과정 중 배급사컬럼 1샘플만 기타 처리, 나머지는 검색된 데이터 삽입 및 사유있는 삭제
#삭제 사유들 : 개봉일기준 2004년 전 영화, 포르노, 단편영화, 독립 및 예술영화,컬럼이 3개이상 비어 있는 샘플, 
# 아이돌 공연 및 아이돌 다큐(특수한 경우라 이상치처리) ,저예산영화,iptv, ott 선개봉 후 스크린수 1인 영화,

# 향후 처리할 프로세스 : 
# 1. 청불영화 중 포르노영화 검열, 
# 2. 배우 컬럼 NA 중 정당사유(애니메이션이나 다큐 등 같은 경우 정당함) 없는 NA 처리
# 3. 영화명 중복 샘플 처리 : 그룹화 하여 맥스값등 집계함수로 합하기
# 4. 등등 

#. '독립'이란 거대 자본과 시장을 지배하는 배급망으로부터 자유롭다는 뜻이다.
#. 저예산영화는 말 그대로 적은 돈을 투자해 만든 작품을 뜻한다. 
#. 다큐멘터리 영화는 연출이 들어가지 않은 실제 영상 기반한 영화.
#. 상업영화 기준으로 분석을 하는 우리로서는 독립영화와, 흥행의 기준이 차이가나는 저예산영화는 삭제,
#. 다큐멘터리 영화는 대부분 독립영화이나, 범주 정의 자체가 다르기 때문에 무조건 독립영화취급 하기 어려움
```

'독립'이란 거대 자본과 시장을 지배하는 배급망으로부터 자유롭다는 뜻이다. 

저예산영화는 말 그대로 적은 돈을 투자해 만든 작품을 뜻한다. 예산 규모가 잣대다. 일반적으로 순제작비 10억 원 미만이면 저예산영화라고 말한다 관객 1만 명 이상이면 성공했다고 평가한다.

다큐멘터리 영화는 연출이 들어가지 않은 실제 영상 기반.

상업영화 기준으로 분석을 하는 우리로서는 독립영화와, 흥행의 기준이 차이가나는 저예산영화는 삭제,
삭제 기준이 애매한 것이 다큐멘터리 영화이다. 노무현입니다. 라는 영화는 180만이 넘어섰고 대박은 아니지만 어느정도 흥행했다 볼 수 있기 때문에..
