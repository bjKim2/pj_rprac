```r
# 첫번쨰 값 리턴 함수
temp_fun <- function(x){
    y<- x
   return(y)
}
```

```r
# 그룹화 , 컬럼별 집계함수가 다르고, 집계함수가 아닌 함수도 필요 3가지 컬럼종류로 나눠서 그룹화 후 merge 예정
x1 <- aggregate(cbind(누적관객수,매출액점유율,누적매출액,스크린수) ~ 영화명,df,max)
x2 <- aggregate(cbind(개봉일,제작사,배급사,등급,장르,감독,배우) ~ 영화명,df,temp_fun)
x3 <- aggregate(cbind(매출액,관객수,상영횟수) ~ 영화명,df,sum)
```

```r
# 로우수 점검
nrow(x1)
nrow(x2)
nrow(x3)
```

```r
x4 <- merge(x1,x3)
x5 <- merge(x4,x2)
## 머지 완료
```

```r
df2<-x5
```
