```r
num1 <- 2
df$배우 <- as.character(df$배우)
for(i in 1:nrow(df)){
    x1 <- df$배우[i]
    x2 <- unlist(strsplit(x1,","))
    if(length(x2)>= num1 ){
        temp <- x2[1:num1]
    }else{
        temp <- x2
    }
    df$배우[i] = ""
    for(j in 1:length(temp)){
         df$배우[i] = paste(df$배우[i],temp[j],sep = ",")
    }
    df$배우[i]<- substr(df$배우[i],2,nchar(df$배우[i]))
}


```
